```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    "Machine learning is transforming technology",
    "Artificial intelligence is changing the world",
    "I enjoy cooking and baking",
    "Natural language processing is part of AI",
    "Deep learning improves prediction accuracy"
]
```

```python
embeddings = model.encode(sentences)

print("Embedding shape:", embeddings.shape)
```

```python
sim_matrix = cosine_similarity(embeddings)

sim_matrix
```

```python
query = "AI and machine learning"

query_vec = model.encode([query])

scores = cosine_similarity(query_vec, embeddings)[0]

for sent, score in sorted(zip(sentences, scores), key=lambda x: -x[1]):
    print(f"{sent} → {score:.3f}")
```

- Sentence transformers generate contextual embeddings for full sentences.
- Similarity is computed using cosine similarity between embedding vectors.
- Higher score = closer semantic meaning.
