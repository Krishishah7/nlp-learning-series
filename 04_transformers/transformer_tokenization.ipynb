```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
text = "Natural Language Processing is amazing!"
```

```python
tokens = tokenizer.tokenize(text)
tokens
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
token_ids
```

```python
encoded = tokenizer.encode(text)
encoded
```

```python
decoded_text = tokenizer.decode(encoded)
decoded_text
```

```python
encoded_inputs = tokenizer(
    text,
    return_tensors="pt",
    padding=True,
    truncation=True
)

encoded_inputs
```

- This notebook demonstrates how transformer models tokenize text using subword tokenization.
- Text is converted into tokens and token IDs before being passed to the model.
- Special tokens such as [CLS] and [SEP] help transformers understand sentence structure.
