```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
faq_data = [
    {
        "question": "What is machine learning?",
        "answer": "Machine learning is a subset of AI that enables systems to learn from data."
    },
    {
        "question": "What is artificial intelligence?",
        "answer": "Artificial intelligence refers to machines that can perform tasks requiring human intelligence."
    },
    {
        "question": "What is natural language processing?",
        "answer": "Natural language processing is a field of AI that focuses on understanding and processing human language."
    },
    {
        "question": "What is deep learning?",
        "answer": "Deep learning is a type of machine learning that uses neural networks with multiple layers."
    },
    {
        "question": "What is Python used for?",
        "answer": "Python is widely used for data science, machine learning, and web development."
    }
]
```

```python
questions = [item["question"] for item in faq_data]
question_embeddings = model.encode(questions)
```

```python
user_query = "Explain NLP"
query_embedding = model.encode([user_query])
```

```python
similarity_scores = cosine_similarity(query_embedding, question_embeddings)[0]
```

```python
best_match_index = np.argmax(similarity_scores)

best_question = faq_data[best_match_index]["question"]
best_answer = faq_data[best_match_index]["answer"]
best_score = similarity_scores[best_match_index]

print("User Query:")
print(user_query)

print("\nBest Matching Question:")
print(best_question)

print("\nAnswer:")
print(best_answer)

print("\nSimilarity Score:")
print(best_score)
```

```python
results = pd.DataFrame({
    "Question": questions,
    "Similarity Score": similarity_scores
}).sort_values(by="Similarity Score", ascending=False)

results
```

- This notebook implements a simple FAQ search system using sentence embeddings.
- FAQ questions and the user query are converted into vector representations using a sentence transformer model.
- Cosine similarity is used to find the question that is semantically closest to the user query, and the corresponding answer is returned.
