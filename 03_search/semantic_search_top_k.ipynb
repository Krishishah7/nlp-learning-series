```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
documents = [
    "Machine learning improves predictive models",
    "Artificial intelligence is transforming industries",
    "Natural language processing deals with text data",
    "I enjoy cooking food during weekends",
    "Deep learning is a subset of machine learning",
    "Python is widely used in data science"
]
```

```python
doc_embeddings = model.encode(documents)
```

```python
query = "machine learning and AI"
query_embedding = model.encode([query])
```

```python
scores = cosine_similarity(query_embedding, doc_embeddings)[0]
```

```python
k = 3

top_k_indices = np.argsort(scores)[::-1][:k]

for idx in top_k_indices:
    print(f"{documents[idx]} â†’ {scores[idx]:.3f}")
```

```python
results = pd.DataFrame({
    "Document": documents,
    "Similarity Score": scores
}).sort_values(by="Similarity Score", ascending=False)

results
```

- This notebook implements a basic semantic search system.
- Documents and user queries are converted into embeddings using a sentence transformer model.
- Cosine similarity is used to retrieve the top-K most relevant documents based on semantic meaning.
